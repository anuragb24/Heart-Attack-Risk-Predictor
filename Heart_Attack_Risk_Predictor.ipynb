
# ü´Ä Heart Attack Risk Predictor

Cardiovascular diseases, especially **heart attacks**, are one of the leading causes of death worldwide. Early prediction of risk factors can help in preventive care and timely treatment. This project applies **machine learning** to predict whether a patient is at **high risk of a heart attack**, using clinical and demographic data.

### üìÇ Dataset

The dataset (`heart.csv`) contains patient health records with the following features:

* **Age** ‚Üí Patient‚Äôs age
* **Sex** ‚Üí Gender (0 = female, 1 = male)
* **cp** ‚Üí Chest pain type (0 = typical angina, 1 = atypical angina, 2 = non-anginal, 3 = asymptomatic)
* **trtbps** ‚Üí Resting blood pressure (mmHg)
* **chol** ‚Üí Serum cholesterol (mg/dl)
* **fbs** ‚Üí Fasting blood sugar >120 mg/dl (1 = true, 0 = false)
* **rest\_ecg** ‚Üí Resting ECG results (0‚Äì2)
* **thalach** ‚Üí Maximum heart rate achieved
* **exang** ‚Üí Exercise-induced angina (1 = yes, 0 = no)
* **ca** ‚Üí Number of major vessels (0‚Äì3)
* **target** ‚Üí Output variable (0 = less chance, 1 = more chance of heart attack)

The dataset is **clean with no missing values**, and correlation analysis shows features like **chest pain type, maximum heart rate, exercise-induced angina, and number of vessels** are most influential in predicting heart attack risk.

### üîç Methodology

1. **Preprocessing** ‚Äì Dropped redundant columns, standardized features.
2. **Exploratory Data Analysis (EDA)** ‚Äì Correlation heatmaps, feature distributions, and target relationships.
3. **Model Training** ‚Äì Implemented Logistic Regression, Decision Tree, Random Forest, KNN, SVM, and Naive Bayes.
4. **Evaluation** ‚Äì Compared models using accuracy, confusion matrices, precision, recall, and F1-score.

### üìä Results

* **Random Forest & SVM** ‚Üí Best-performing models with highest accuracy and stability.
* **Logistic Regression** ‚Üí Strong baseline, highly interpretable.
* **Decision Tree & Naive Bayes** ‚Üí Simple but less accurate.
* **KNN** ‚Üí Moderate accuracy, sensitive to parameter choice.

### ‚úÖ Conclusion

This study shows that **machine learning models can effectively predict heart attack risk** based on health parameters. Ensemble methods like Random Forest provide robust performance, while Logistic Regression remains valuable for clinical interpretability. Such predictive systems can serve as **decision-support tools** for early detection and risk assessment in healthcare.

### K Nearest Neighbour

#### We have to select what k we will use for the maximum accuracy

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate= []
for i in range(1,40):
    knn= KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,encoded_y)
    pred= knn.predict(x_test)
    error_rate.append(np.mean(pred != encoded_ytest))
```
